```python
import pandas as pd
import numpy as np
import pathlib as path
import folium
import geopandas as gpd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('database.csv')
df.head()
```

```python
df.info()
```

```python
df.describe()
```

```python
df2 = df.copy()

#dropping columns with missing values
missing_values_columns = [col for col in df2.columns
                     if df2[col].isnull().any()]
df2 = df2.drop(missing_values_columns, axis=1)
df2.head()
```

```python
lengths = df2["Date"].str.len()
lengths.value_counts()

#having a look at the fishy datapoints
wrongdates = np.where([lengths == 24])[1]
print("Fishy dates:", wrongdates)
df2.loc[wrongdates]

#fixing the wrong dates and changing the datatype from numpy object to datetime64[ns]
df2.loc[3378, "Date"] = "02/23/1975"  
df2.loc[7512, "Date"] = "04/28/1985"
df2.loc[20650, "Date"] = "03/13/2011"
df2['Date']= pd.to_datetime(df2["Date"])
df2.info()
```

```python
#exploring the length of time objects
lengths = df2["Time"].str.len()
lengths.value_counts()

#Having a look at the fishy datapoints
wrongtime = np.where([lengths == 24])[1]
print("Fishy time:", wrongtime)
df2.loc[wrongtime]

#fixing the wrong time and changing the datatype from numpy object to timedelta64[ns]
df2.loc[3378, "Time"] = "02:58:41"
df2.loc[7512, "Time"] = "02:53:41"
df2.loc[20650, "Time"] = "02:23:34"
df2['Time']= pd.to_timedelta(df2['Time'])
df2.info()
```

```python
plt.figure(figsize=(12,8))
sns.lineplot(x=df2['Date'].dt.year, y='Magnitude',
             hue='Type', data=df2)
```

```python
gdf = gpd.GeoDataFrame(df2, geometry=gpd.points_from_xy(df["Longitude"], df["Latitude"]))

plt.figure(figsize=(20, 20))
gdf.plot(color="red", marker="o")
plt.show()
```

```python
df2['Location Source'].unique()
```

```python
plt.figure(figsize=(20,20))
sns.barplot(x='Location Source', y='Magnitude', hue='Type', data=df2)
```

```python
sns.distplot(df2['Depth'])
```

```python
# Create a Folium map
m = folium.Map(tiles="cartodbpositron", prefer_canvas=True, zoom_start=5)

# Add the points to the map
for lat, lon in zip(df["Latitude"], df["Longitude"]):
    folium.Marker([lat, lon]).add_to(m)

# Display the map
m.save("map.html")
```

# India's Earthquake Data

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('Official Website of National Center of Seismology.xlsx', header=1)
df
```

```python
df.info()
```

## Without Time-Series

```python
df2 = df.drop(columns=['Location'], axis=1)
df2.head(10)
```

```python
df2.shape
```

```python
df2['Origin Time'] = pd.to_datetime(df2['Origin Time'])
df2
```

### Visualization

```python
sns.pairplot(df2, hue='Magnitude')
```

```python
sns.relplot(x='Origin Time', y='Magnitude', data=df2, kind='line')
```

```python
sns.relplot(x='Depth', y='Magnitude', data=df2, kind='line', legend='full')
```

```python
sns.violinplot(x='Magnitude', y='Depth', data=df2)
```

```python
sns.boxplot( x='Depth', y='Magnitude', data=df2)
```

```python
sns.scatterplot(x='Lat', y='Long', size='Magnitude', hue='Magnitude', data=df2)
```
